```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.base import clone
```

```python
dataset = data.iloc[:,1:]
dataset.head()
```

```python
 dataset.info()
```

```python
category = 'Educacion'                                
data_temp =  data.groupby([category]).agg({ 'Id': "count"})
data_temp = data_temp.reset_index()
data_temp
```

```python
dataset['Trabajo'] = dataset['Trabajo'].astype('category')
dataset['Estado Civil'] = dataset['Estado Civil'].astype('category')
dataset['Educacion'] = dataset['Educacion'].astype('category')
dataset['Credito por Defecto'] = dataset['Credito por Defecto'].astype('category')
dataset['Prestamo Vivienda'] = dataset['Prestamo Vivienda'].astype('category')
dataset['Prestamo Personal'] = dataset['Prestamo Personal'].astype('category')
dataset['Contacto'] = dataset['Contacto'].astype('category')
dataset['Mes'] = dataset['Mes'].astype('category')
dataset['Dia Semana'] = dataset['Dia Semana'].astype('category')
dataset['Resultado Campaña Anterior'] = dataset['Resultado Campaña Anterior'].astype('category')
```

```python
#Separamos la data, en variables independientes (x) y dependientes (y), para poder entrenar un árbol de clasificación
x = dataset.drop("Subscripcion Deposito", axis=1)
y = dataset["Subscripcion Deposito"]

#Mediante el método "train_test_split" usaremos el 20% de la data para probar el modelo.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)

print(f"Train shape: {x_train.shape}")
print(f"Test shape: {x_test.shape}")
```

```python
#Entrenamos el random forest de calsificasion
model = RandomForestClassifier(criterion='entropy', random_state=42)
model.fit(x_train, y_train)
```
